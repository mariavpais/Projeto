### Tratamento dos dados: conversão da hora num formato datetime correto

```python
import datetime
dados_acidentes["Datahora"]=dados_acidentes['Datahora'].apply(lambda x: datetime.datetime.strptime(x, '%Y:%m:%d %H:%M:%S') if type(x)==str else np.NaN)
```

```python
dados_acidentes['Data'] = dados_acidentes['Datahora'].map(lambda x: str(x)[:-8])
```

```python
dados_acidentes["Data"] = pd.to_datetime(dados_acidentes["Data"])
```

```python
dados_acidentes["Mês"]=dados_acidentes["Datahora"].dt.month
```

```python
dados_acidentes["Dia"]=dados_acidentes["Datahora"].dt.day
```

```python
dados_acidentes["Hora"]=dados_acidentes["Datahora"].dt.hour
```

```python
dados_acidentes["Ano"] = dados_acidentes["Datahora"].dt.year
```

```python
dados_acidentes.shape
```

```python
del dados_acidentes["Datahora"]
```

### Tratamento dos dados: conversão das variáveis Latitude e Longitude

```python
dados_acidentes["Latitude GPS"]=dados_acidentes["Latitude GPS"].str.replace(',','.')
```

```python
dados_acidentes["Longitude GPS"]=dados_acidentes["Longitude GPS"].str.replace(',','.')
```

```python
dados_acidentes["Latitude GPS"]=dados_acidentes["Latitude GPS"].apply(lambda x: float(x))
dados_acidentes["Longitude GPS"]=dados_acidentes["Longitude GPS"].apply(lambda x: float(x))
```
